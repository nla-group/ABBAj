```julia
include("ABBA.jl")
using Plots, .ABBA
```

```julia
ts = load_samples()
```

```julia
symbols, model = fit_transform(ts, 5, 0.1, 1);
r_time_series = inverse_transform(symbols, model, ts[1]);
```

```julia
plotlines = hcat(ts, r_time_series);
plot_img = plot(plotlines, label = ["time series" "ABBA reconstruction"], lw = 3);
display(plot_img);
savefig(plot_img,"demo.png")
```
